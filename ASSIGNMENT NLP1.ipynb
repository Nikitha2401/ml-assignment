```python

import nltk
import re
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer,LancasterStemmer,SnowballStemmer,WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sys import getsizeof
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
stp=stopwords.words('english')
print(stp)
```

```python
df=pd.read_csv(r"C:\Users\Venkatesham\nyka_top_brands_cosmetics_product_reviews.zip")
df
```

```python
df.fillna({'review_text':'very good'},inplace=True)
df.isnull().sum()
```

```python
lf=[]
for sent in df['review_text']:
    l=[]
    ls=str(sent).lower()
    ls=re.sub('[^a-z]',' ',ls)
    for word in word_tokenize(ls):
        if word in stp:
            pass
        else:
            stems=PorterStemmer().stem(word)
            l.append(stems)
    lf.append(' '.join(l))
```

```python
lf
```

```python
pd.DataFrame({"cleaned_data":lf})
```

```python
lf=[]
for sent in df['review_text']:
    l=[]
    ls=str(sent).lower()
    ls=re.sub('[^a-z]',' ',ls)
    for word in word_tokenize(ls):
        if word in stp:
            pass
        else:
            stems=SnowballStemmer("english").stem(word)
            l.append(stems)
    lf.append(' '.join(l))
```

```python
lf=[]
for sent in df['review_text']:
    l=[]
    ls=str(sent).lower()
    ls=re.sub('[^a-z]',' ',ls)
    for word in word_tokenize(ls):
        if word in stp:
            pass
        else:
            lemmatize=WordNetLemmatizer().lemmatize(word)
            l.append(lemmatize)
    lf.append(' '.join(l))
```

```python
lf
```

```python
df1=pd.DataFrame({"cleaned_data":lf})
df1
```

```python
cv=CountVectorizer()
nd=cv.fit_transform(df1["cleaned_data"])
nd.toarray()
```

```python
pd.DataFrame(nd.toarray(),columns=sorted(cv.vocabulary_.keys()))
```
